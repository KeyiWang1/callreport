```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
call_report = pd.read_csv('C:/Users/kwang648/Downloads/banking/call_report.csv')
```

```python
call_report['DATE'] = pd.to_datetime(call_report['DATE'],format='%Y-%m-%d')
call_report=call_report.sort_values(by=['DATE'])
```

```python
plt.plot(call_report['DATE'].unique(),call_report.groupby('DATE')['liquidity'].mean())
```
